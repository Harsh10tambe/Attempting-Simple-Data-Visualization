```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Second Hand Car dataset.csv")
```

```python
df.head()
```

#### This data is downloaded from kaggle: https://www.kaggle.com/mayankpatel14/second-hand-used-cars-data-set-linear-regression
#### The data consists a list of price and specifications of second-hand used cars. Let's say we wish to buy a seconf hand car, and we are calculating how much budget to have. So we will use this data to draw a picture to understand how much the price of the car changes with respect to the distance it has been driven for. 

#### Let's make a raw image first. 

```python
num = 50
x = df['km']
y = df['current price']
plt.scatter(x, y)
```

#### The above graph is a scatter plot of 'price' against the 'distance driven in KMs' driven

#### Lets us make it visually appealing and easier to understand.

```python
# Making it black to look better.
# Changin the alpha value, so that overlapping points appear darket in color.
# Reducing the size so that each dot is clearly visible.
plt.scatter(x, y, c='Black', alpha=0.4, s=7)


# Setting the title
plt.title('Second hand price \n v/s \n KMs Driven')


# Naming the axises
plt.xlabel('Price')
plt.ylabel('KMs driven')


# Changing the limits so that the graph looks well fitted on the chart.
# flipping the limits in x-axis so that the graph is flipped horizontally and is easier to interpret.
plt.xlim(150000,50000)
plt.ylim(-10000,650000)

# Printing the graph
plt.show()
```

### By eyeballing, it can ve observed that we can buy a car that is 20,000 dollars cheaper if it driven by 150000 KMs more.
